# 2. 데이터 형태일치

### 2.1 tripdavisor 특수 경우

```python
tripadvisor_except.loc[tripadvisor_except.날짜 == "Hwang이 리뷰를 작성했습니다"]
```

```python
from datetime import datetime
from datetime import date
date_format = "%Y년 %m월"
date_list = []
for text in tripadvisor_except["날짜"]:
    try :
        new_date = datetime.strptime(text,date_format)
        date_list.append(new_date)
    except ValueError:
        date_list.append(None)

tripadvisor_except["날짜2"] = pd.Series(date_list)
```

```python
tripadvisor_except.iloc[190,:]
```

```python
tripadvisor_except.iat[190,2] = None
```

```python
tripadvisor_except.iloc[190,:]
```

```python
new_tripadvisor_except = pd.concat([tripadvisor_except.장소,tripadvisor_except.평점,tripadvisor_except.날짜,tripadvisor_except.리뷰], axis = 1)
```

```python
new_tripadvisor_except["작성연도"] = tripadvisor_except.날짜2.dt.year.astype("Int64")
new_tripadvisor_except["작성월"] = tripadvisor_except.날짜2.dt.month.astype("Int64")
new_tripadvisor_except["작성일"] = np.NaN
new_tripadvisor_except["평점"] = tripadvisor_except.평점.astype("Int64")
```

```python
new_tripadvisor_except = new_tripadvisor_except[["장소","평점","날짜","작성연도","작성월","작성일","리뷰"]]
new_tripadvisor_except
```

### 2.2 tripadvisor 일반 경우

```python
tripadvisor_normal["date"] = tripadvisor_normal["date"].astype(str).astype("datetime64[ns]")
```

```python
new_tripadvisor_normal = pd.concat([tripadvisor_normal.location_name,tripadvisor_normal.rate,tripadvisor_normal.date,tripadvisor_normal.review],axis = 1)    
new_tripadvisor_normal["작성연도"] = new_tripadvisor_normal.date.dt.year.astype(int)
new_tripadvisor_normal["작성월"] = new_tripadvisor_normal.date.dt.month.astype(int)
new_tripadvisor_normal["작성일"] = new_tripadvisor_normal.date.dt.day
new_tripadvisor_normal["rate"] = new_tripadvisor_normal.rate.astype(int)
```

```python
new_tripadvisor_normal.columns = ["장소","평점","날짜","리뷰","작성연도","작성월","작성일"]
```

```python
new_tripadvisor_normal = new_tripadvisor_normal[["장소","평점","날짜","작성연도","작성월","작성일","리뷰"]]
new_tripadvisor_normal
```

### 2.3 trip.com 리뷰

```python
tripcom
```

```python
from datetime import datetime
from datetime import date
date_format = "%Y년%m월%d일"
date_list = []
for text in tripcom["리뷰작성일"]:
    new_date = datetime.strptime(text,date_format)
    date_list.append(new_date)

tripcom["날짜"] = pd.Series(date_list)
tripcom
```

```python
new_tripcom = pd.concat([tripcom.장소,tripcom.평점,tripcom.날짜,tripcom.리뷰],axis =1)
new_tripcom["작성연도"] = new_tripcom.날짜.dt.year
new_tripcom["작성월"] = new_tripcom.날짜.dt.month
new_tripcom["작성일"] = new_tripcom.날짜.dt.day
new_tripcom = new_tripcom[["장소","평점","날짜","작성연도","작성월","작성일","리뷰"]]
```
